```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
plt.plot(x,y)
plt.title("REAL GOAT")
plt.xlabel("KELP BOTIYA")
plt.ylabel("RAHUL PRO")
```

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.title("REAL GOAT")
plt.xlabel("KELP BOTIYA")
plt.ylabel("RAHUL PRO")
plt.subplot(2,2,3)
plt.subplot(2,2,4)
```

```python
fig=plt.figure(figsize=(10,12),dpi=100)
axes1=fig.add_axes([.4,.5,1,.5])
axes1.plot(x,y)
axes2=fig.add_axes([.7,.5,.3,.3])
axes2.plot(x,y)
```
