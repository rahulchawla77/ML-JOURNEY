```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('/kaggle/input/datasets/shambhurajejagadale/student-performance-prediction-dataset/student_dataset_10000_rows.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['placement_status']=le.fit_transform(df['placement_status'])
```

```python
 from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
df_scaled= pd.DataFrame(
    scaler.fit_transform(df),
    columns=df.columns
)
```

```python
X=df_scaled.drop(columns=['placement_status','exam_score'])
y=df['placement_status']
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score=cross_val_score(model,X,y,cv=5,scoring='accuracy')
```
